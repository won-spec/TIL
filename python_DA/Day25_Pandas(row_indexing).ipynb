```python
#분석용 함수 (기본통계함수)
#평균, 편차, 분산, 표준편차, 공분산, 상관계수
#평균(mean)
#편차(deviation): 확률변수 x와 mean의 차이   편차의 합은 0
#분산(variance):데이터의 흩어짐 정도를 알기위해서 사용하는 값으로 편차의 제곱의 평균
# 표준편차(standard deviation):분산의 제곱근   
#     분산의 단위 문제를 해결하고자 사용 
import numpy as np
arr=np.array([4,6,1,3,8,8],dtype=np.int32)
print(arr)
print(arr.sum())   #합계
print(np.sum(arr))
print(arr.mean())
print(arr.var())  #분산
print(arr.std())#표준편차
```

```python
#공분산
# 공분산: 두개의 확률변수 X,Y가 있을때
#     이 두개의 확률변수의 관계를보여줄 때 사용해요
# 공분산은 두 확률변수의 편차의 곱의 평균으로 표현
#데이터가 평균으로부터 얼마나 떨어져있느냐를 표현 
#그래프를 그려보아요
import numpy as np
import matplotlib.pyplot as plt
#만약 module이 없다고 error가 뜨면 conda install matplotlib
np.random.seed(1)  #항상 같은 난수가 발생하도록 seed 처리 
x=np.random.randint(-20,20,(10,))
y=np.random.randint(-10,10,(10,))
x_mean=x.mean()
y_mean=y.mean()
x_deviation=x-x_mean
y_deviation=y-y_mean
print(x_deviation)
print(y_deviation)

result=0
for i in range(10):
    result+=(x_deviation[i]*y_deviation[i])
myCov=result/9   #모공분산: 원래공식n으로나눔    
# 표준공분산:(n-1)로 나누어줘요=>추정치가 좋아져요 

print(myCov)
print(np.cov(x,y))
#scatter(산점도) 표현해보아요
plt.scatter(x,y,color="red")
plt.scatter(x_mean,y_mean,color="blue")
plt.show()
```
